```python
%matplotlib inline
# Dependencies and Setup
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# File to Load 
file_to_load = "Resources/fire_Final.csv"

# Read Purchasing File and store into Pandas data frame
fire_data = pd.read_csv(file_to_load)

fire_data.head()
```

```python
#Select only 2005
fire_2005 = fire_data[fire_data.FIRE_YEAR == 2005]
#group by State
state_2005 = fire_2005.groupby(["STATE"])
#calculate total Fire_zize per state
state_fire_size_2005 = state_2005["FIRE_SIZE"].sum()
#format and write to a new csv
state_fire_size_2005_df = pd.DataFrame(state_fire_size_2005)
state_fire_size_2005_df.style.format({"FIRE_SIZE":'{:,.2f}'})
state_fire_size_2005_df.rename(columns={'FIRE_SIZE':'FIRE_SIZE_2005'}, inplace=True)
state_fire_size_2005_df.to_csv("RESOURCES/fire_size_2005.csv")
state_fire_size_2005_df.head(20)
```

```python
#Select only 2002
fire_2002 = fire_data[fire_data.FIRE_YEAR == 2002]
#group by State
state_2002 = fire_2002.groupby(["STATE"])
#calculate total Fire_zize per state
state_fire_size_2002 = state_2002["FIRE_SIZE"].sum()
#format and write to a new csv
state_fire_size_2002_df = pd.DataFrame(state_fire_size_2002)
state_fire_size_2002_df.style.format({"FIRE_SIZE":'{:,.2f}'})
state_fire_size_2002_df.rename(columns={'FIRE_SIZE':'FIRE_SIZE_2002'}, inplace=True)
state_fire_size_2002_df.to_csv("RESOURCES/fire_size_2002.csv")
state_fire_size_2002_df.head(15)
```

```python
#Select only 2011
fire_2011 = fire_data[fire_data.FIRE_YEAR == 2011]
#group by State
state_2011 = fire_2011.groupby(["STATE"])
#calculate total Fire_zize per state
state_fire_size_2011 = state_2011["FIRE_SIZE"].sum()
#format and write to a new csv
state_fire_size_2011_df = pd.DataFrame(state_fire_size_2011)
state_fire_size_2011_df.style.format({"FIRE_SIZE":'{:,.2f}'})
state_fire_size_2011_df.rename(columns={'FIRE_SIZE':'FIRE_SIZE_2011'}, inplace=True)
state_fire_size_2011_df.to_csv("RESOURCES/fire_size_2011.csv")
state_fire_size_2011_df.head(10)
```

```python
state_acres_df = pd.read_csv("RESOURCES/states_acres.csv")
```

```python
#Import the  acres per state data
state_acres_df = pd.read_csv("RESOURCES/states_acres.csv")

#replace object to float do Totat Acres
state_acres_df['Total acres'] = state_acres_df['Total acres'].apply(lambda x: float(x.split()[0].replace(',', '')))
state_acres_df['Total acres'] = state_acres_df['Total acres'].astype(float)

# Combine the data into a single dataset

state_fire_size_3ya_df = pd.merge(state_acres_df, state_fire_size_2002_df, how="left", on="STATE")
state_fire_size_3yb_df = pd.merge(state_fire_size_3ya_df, state_fire_size_2005_df, how="left", on="STATE")
state_fire_size_3yc_df = pd.merge(state_fire_size_3yb_df, state_fire_size_2011_df, how="left", on="STATE")

state_fire_size_3y_df = state_fire_size_3yc_df.dropna(how="any")

state_fire_size_3y_df.style.format({"FIRE_SIZE_2011":'{:,.2f}'})
state_fire_size_3y_df.style.format({"FIRE_SIZE_2005":'{:,.2f}'})
state_fire_size_3y_df.style.format({"FIRE_SIZE_2002":'{:,.2f}'})

#print(state_fire_size_3y_df.dtypes)
state_fire_size_3y_df.head(50)
```

```python
state_percent_2002 = (state_fire_size_3y_df["FIRE_SIZE_2002"]/ state_fire_size_3y_df["Total acres"])*100
state_percent_2005 = (state_fire_size_3y_df["FIRE_SIZE_2005"]/ state_fire_size_3y_df["Total acres"])*100
state_percent_2011 = (state_fire_size_3y_df["FIRE_SIZE_2011"]/ state_fire_size_3y_df["Total acres"])*100



total_fire_3y_n = pd.concat([state_fire_size_3y_df, state_percent_2002.rename("Percent 2002"), state_percent_2005.rename("Percent 2005"), state_percent_2011.rename("Percent 2011")],  axis = 1)

#format_dict = {'Percent 2002': '{:.2%}', 'Percent 2005': '{:.2%}', 'Percent 2011': '{:.2%}'}
#total_fire_3y_n.style.format(format_dict).hide_index()

#total_fire_3y_n.sort_values(["Total acres"], ascending = False).style.format({"Percent 2002":"{:.2%}"})

#print(total_fire_3y_n.dtypes)
total_fire_3y_n.style.format({"FIRE_SIZE_2011":'{:,.2%}'})
total_fire_3y_n.style.format({"FIRE_SIZE_2005":'{:,.2%}'})
total_fire_3y_n.style.format({"FIRE_SIZE_2002":'{:,.2%}'})
total_fire_3y_n
```
